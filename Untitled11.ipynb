```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv("/content/mobile_price_range_data (2).csv")
df
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
from sklearn.utils import shuffle
ndf = shuffle(df,random_state=42)
ndf
```

```python
print(df.info())
```

```python
ndf = ndf[ndf['sc_w']!=0]
ndf
```

```python
price_plot=ndf['price_range'].value_counts().plot(kind='bar')
plt.xlabel('price_range')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(ndf.corr(),annot=True,cmap=plt.cm.Accent_r)
plt.show()
```

```python
y = ndf['price_range']
x = ndf.drop('price_range', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
y_train
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def my_confusion_matrix(y_test, y_pred, plt_title):
    cm=confusion_matrix(y_test, y_pred)
    print(classification_report(y_test, y_pred))
    sns.heatmap(cm, annot=True, fmt='g', cbar=False, cmap='BuPu')
    plt.xlabel('Predicted Values')
    plt.ylabel('Actual Values')
    plt.title(plt_title)
    plt.show()
    return cm
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='newton-cg',multi_class='auto')
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
lr_acc = accuracy_score(y_test,y_pred_lr)
```

```python
print('LogisticRegression accuracy score: ',lr_acc)
cm_rfc=my_confusion_matrix(y_test, y_pred_lr, 'LogisticRegression Confusion Matrix')
```

```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred_dt = dt.predict(x_test)
dt_acc= accuracy_score(y_test,y_pred_dt)
```

```python
print('Decision tree Accuracy Score: ',dt_acc)
cm_rfc=my_confusion_matrix(y_test, y_pred_dt, 'Decision tree Confusion Matrix')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(bootstrap= True,
                           max_depth= 7,
                           max_features= 15,
                           min_samples_leaf= 3,
                           min_samples_split= 10,
                           n_estimators= 200,
                           random_state=7)
```

```python
rfc.fit(x_train, y_train)
y_pred_rfc=rfc.predict(x_test)
rtc_acc=accuracy_score(y_test,y_pred_rfc)
```

```python
print('Random Forest Classifier Accuracy Score: ',rtc_acc)
cm_rfc=my_confusion_matrix(y_test, y_pred_rfc, 'Random Forest Confusion Matrix')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3,leaf_size=25)

knn.fit(x_train, y_train)
y_pred_knn=knn.predict(x_test)
knn_acc = accuracy_score(y_test,y_pred_knn)
print('KNN Classifier Accuracy Score: ',knn_acc)
cm_rfc=my_confusion_matrix(y_test, y_pred_knn, 'KNN Confusion Matrix')
```

```python
from sklearn import svm
svm_clf = svm.SVC(decision_function_shape='ovo')
svm_clf.fit(x_train, y_train)
y_pred_svm=svm_clf.predict(x_test)
svm_acc = accuracy_score(y_test,y_pred_svm)
print('SVM Classifier Accuracy Score: ',svm_acc)
cm_rfc=my_confusion_matrix(y_test, y_pred_svm, 'SVM Confusion Matrix')
```

```python
plt.bar(x=["lr","dt","rtc","knn","svm"],height = [lr_acc,dt_acc,rtc_acc,knn_acc,svm_acc])
x.lable = "Alogrithm"
y.lable = "accuracy score"
plt.show()
```
